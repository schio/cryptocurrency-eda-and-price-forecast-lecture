```python
import requests
import pandas as pd
import json
```

```python
url = "https://api.upbit.com/v1/candles/weeks?market=KRW-BTC&count=200"

headers = {"accept": "application/json"}

response = requests.get(url, headers=headers)

btc_df = pd.DataFrame(json.loads(response.text))
btc_df
```

```python
weeks_path = "/content/drive/MyDrive/coin_data/weeks"
```

```python
btc_df.to_csv(weeks_path + "/krw_btc.csv", index=False)
```
