# 비트코인의 주 단위 캔들 정보를 최대한 과거 데이터 까지 가져온 후 시간순으로 정렬하기

```python
import time
import requests
import pandas as pd
import json
```

```python
datetimes = []
for year in range(2018,2025):
    datetimes.append(f"{year}-01-01T00%3A00%3A00%2B09%3A00")
datetimes
```

```python
headers = {"accept": "application/json"}
dfs = []
for datetime in datetimes:
    url = f"https://api.upbit.com/v1/candles/weeks?market=KRW-BTC&to={datetime}&count=53"
    response = requests.get(url, headers=headers)
    time.sleep(0.08)
    dfs.append(pd.DataFrame(json.loads(response.text)))
```

```python
df = pd.concat(dfs)
```

```python
df = df.drop(columns=["candle_date_time_utc", "timestamp", "first_day_of_period"])
df.columns = [
    "market",
    "datetime",
    "open",
    "high",
    "low",
    "close",
    "trade_price",
    "trade_volume",
]
```

```python
df = df.sort_values(by="datetime").reset_index(drop=True)
df = df.drop_duplicates()
df
```
