# DataFrame의 기능 알아보기

```python
import pandas as pd
titanic_df = pd.read_csv("titanic.csv")
```

#### 1개 열 추출하기

```python
titanic_df["age"]
```

#### N개 열 추출하기

```python
titanic_df[["age", "sex", "fare"]]
```

#### 상위 일부 행 확인하기

```python
titanic_df.head()
```

```python
titanic_df.head(30)
```

#### 간단한 정보 확인하기

```python
titanic_df.info()
```

#### 결측치 확인하기

```python
titanic_df.isna()
```

#### 각 열의 결측치 개수 확인하기

```python
titanic_df.isna().sum()
```

#### 열 삭제하기

```python
titanic_df = titanic_df.drop(columns=["alive", "class"])
titanic_df
```

```python
titanic_df.drop(columns=["asdf"])
```

#### 비교연산자를 활용하여 조건으로 검색하기

```python
titanic_df[titanic_df["sex"] == "male"]
```

```python
titanic_df[titanic_df["age"] > 20]
```

```python
titanic_df[titanic_df["survived"] == 0]
```

#### isin() 활용하여 조건으로 검색하기

```python
titanic_df[titanic_df["who"].isin(["man", "child"])]
```

#### 통계치 확인하기

```python
titanic_df.max(numeric_only=True)
```

```python
titanic_df.min(numeric_only=True)
```

```python
titanic_df.mean(numeric_only=True)
```

```python
titanic_df.std(numeric_only=True)
```

```python
titanic_df.count()
```

```python
titanic_df.describe()
```

#### 그룹핑하여 통계치 확인하기

```python
titanic_df.groupby(["sex"])["fare"].mean()
```

```python
titanic_df.groupby(["sex"])["survived"].mean()
```

```python
titanic_df.groupby(["survived"])["fare"].mean()
```

#### 오름차순 정렬하기

```python
titanic_df.sort_values(by=["fare"])
```

#### 내림차순 정렬하기

```python
titanic_df.sort_values(by=["fare"], ascending=False)
```
