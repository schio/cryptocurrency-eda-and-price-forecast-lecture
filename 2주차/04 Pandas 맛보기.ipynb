# Numpy 맛보기

```python
import numpy as np
```

```python
my_list = [10, 20, 30, 40, 50]
print(my_list)
print(type(my_list))
```

```python
my_array = np.array(my_list)
print(my_array)
print(type(my_array))
```

```python
print(my_array.min())
print(my_array.max())
print(my_array.mean())
print(my_array.std())
```

```python
print(np.min(my_array))
print(np.max(my_array))
print(np.mean(my_array))
print(np.std(my_array))
```

# Pandas

#### Series 선언하기

```python
import pandas as pd
```

```python
my_list = [10, 20, 30, 40, 50]
my_series = pd.Series(my_list)
my_series
```

#### Series 기능 알아보기

```python
my_list = [10, 20, 30, 40, 50]
my_series = pd.Series(my_list)
my_series
```

```python
print(my_series.values)
print(my_series.values[1])
print(my_series.values[:3])
print(my_series.values.min())
print(my_series.values.max())
```

**Series간의 연산**

```python
husband_income = pd.Series(
    [300, 300, 400, 300, 300, 400, 300, 300, 400, 300, 300, 400]
)

wife_income = pd.Series(
    [400, 200, 600, 400, 200, 600, 400, 200, 600, 400, 200, 600]
)

family_income = husband_income + wife_income
family_income
```

#### DataFrame 선언하기
